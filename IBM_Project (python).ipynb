# Load Dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv(r"C:\Users\acer\Downloads\student_dataset (1).csv")
df.head()
```

# ðŸ‘‰ðŸ»Data Cleaning ( Using pandas)

```python
# for missing values(give true)
df.isnull()  
# to remove missing values(NaN,none)
df.dropna()  
# true for duplicates
df.duplicated()     
# remove duplicates
df.drop_duplicates()   
# fill 0 where value is null  
df.fillna(0)
# count null values
df.isnull().sum()

```

**Create pass\fail column**

```python
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 70 else "Fail")
print(df["Result"])
```

# Data Visualization ( With Matplotlib and Seaborn)

**ðŸ‘‰ðŸ»Pie Chart -> (Pass vs Fail)**

```python
pass_count = (df["Marks"] >= 70).sum()
fail_count = (df["Marks"] < 70).sum()

labels = ["Pass", "Fail"]
sizes = [pass_count, fail_count]

plt.figure(figsize=(12,6))
plt.pie(sizes, labels=labels, autopct="%1.1f%%",shadow=True, startangle=90)
plt.title("Pass vs Fail")
plt.show()
```

**Summary** :

âœ”The code calculates how many students passed and failed based on marks:

âœ”Pass: Marks â‰¥ 70

âœ”Fail: Marks < 70

**ðŸ‘‰ðŸ»Bar plot -> Student-wise marks**

```python
plt.figure(figsize=(12,8))
sns.countplot(data=df,x="Grade",color="green",hue="Grade",edgecolor="Black")
plt.show()
```

**Summary** :

âœ”The code uses Seaborn's countplot to show how many students are in each Grade category. 

âœ”The plot helps you quickly see which grade has the highest or lowest frequency.

**Histogram for Marks Distribution**

```python
plt.figure(figsize=(10,6))
sns.histplot(df["Marks"], bins=20, kde=True,color="pink",edgecolor="black", alpha=1.0)
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

**Summary**:

âœ”This code creates a histogram showing how studentsâ€™ marks are distributed.

âœ”It divides the marks into 20 bins (intervals) and displays how many students fall into each interval.

âœ”It shows how marks are spread across all students.

**ðŸ‘‰ðŸ»Scatter Plot ->Attendance Vs Marks**

```python
plt.figure(figsize=(12,5))
plt.scatter(df["Attendance"], df["Marks"],marker='*',label="Student data",s=70,color="red")
plt.title("Attendance vs Marks")
plt.xlabel("Attendance")
plt.ylabel("Marks")
plt.grid(True)
plt.legend()
plt.show()
```

**Summary** :

âœ”This code creates a scatter plot to show the relationship between Attendance (%) and Marks of students.

âœ”Each point (*) represents a student.

âœ”It helps you visually check whether students with higher attendance generally score higher marks.


**ðŸ‘‰ðŸ»Violin Plot -> Distribution of Marks**

```python
plt.figure(figsize=(8,6))
sns.violinplot(y=df["Marks"], palette="dark")
plt.title("Distribution of Marks")
plt.show()
```

**Summary:**

âœ” This code creates a violin plot to show the full distribution of Marks.
It displays:

âœ” The spread of marks

âœ” The density (where most marks are concentrated)

âœ” A smooth shape showing how marks are distributed from low to high

ðŸ‘‰ It visually shows where most studentsâ€™ marks lie and how the marks vary.

**ðŸ‘‰ðŸ»Box Plot of Marks**

```python
plt.figure(figsize=(10,5))
sns.boxplot(y=df["Marks"],vert=True, patch_artist=True,boxprops=dict(facecolor="cyan"))
plt.title("Boxplot of Student Marks")
plt.ylabel("Marks")
plt.show()
```

**Summary**:

âœ” It gives a quick overview of mark distribution and identifies extreme values.

âœ” Minimum marks

âœ” 25th percentile (bottom edge)

âœ” Median (middle value)

âœ” 75th percentile (upper edge)

âœ” Maximum marks

Any outliers (points far from the rest of the data)

**ðŸ‘‰ðŸ»Heatmap of Student Data**

```python
df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10,5))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="BrBG")

plt.title("Correlation Heatmap of Student Dataset")
plt.xlabel("Features")
plt.ylabel("Features")

plt.show()
```

**Summary:**

âœ” 1. Find relationships between columns

âœ” 2. Understand data patterns quickly

âœ” 3. Identify strong & weak correlations

ðŸ‘‰ Strong relationship = darker/bright color
ðŸ‘‰ Weak relationship = lighter color
